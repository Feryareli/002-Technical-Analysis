<h2 style="color:#2e86c1; background-color:#f0f0f0; padding:10px; border-radius:10px; text-align:center;">Project Overview</h2>

<p style="color:#2e86c1;">
  This project focuses on developing and optimizing trading strategies using historical price data from <strong>Apple Inc. (AAPL)</strong> and <strong>Bitcoin (BTC-USD)</strong>. The main objective is to maximize returns by leveraging five key technical indicators: RSI, Bollinger Bands, MACD, ATR, and Williams %R. 
</p>

<p style="color:#2e86c1;">
  The strategy employs a rigorous approach where buy/sell signals are defined for each indicator, and all possible combinations of indicators are tested (2^5 - 1 combinations). Backtesting is performed for both long and short positions while tracking cash and portfolio value over time. The performance of each strategy is evaluated based on <strong>Sharpe Ratio</strong> (with treasury bonds as a reference), <strong>Max Drawdown</strong>, and <strong>Win-Loss Ratio</strong>.
</p>

<p style="color:#2e86c1;">
  The project leverages <strong>Optuna</strong> for parameter optimization, tuning variables such as stop-loss, take-profit, indicator windows, and the number of shares traded. 
</p>

<p style="color:#2e86c1;">
  After selecting the optimal strategy based on training data, it is applied to a test dataset for validation, comparing it against a passive buy-and-hold strategy. In this Jupyter notebook report, you will find the visualizations comparing the performance of the active and passive strategies for both Apple and Bitcoin datasets. The report highlights how each strategy behaved, including the optimized parameters used for the active strategy, as well as the final capital obtained in each case. Key metrics such as Sharpe Ratio, Max Drawdown, and Win-Loss Ratio are also included to provide a comprehensive view of the strategies' effectiveness. </p>
</p>

<h2 style="color:#2e86c1; background-color:#f0f0f0; padding:10px; border-radius:10px; text-align:center;">Indicator Selection</h2>


<p style="color:#2e86c1;">
  The selected indicators for this trading strategy are <strong>RSI</strong>, <strong>Bollinger Bands</strong>, <strong>MACD</strong>, <strong>ATR</strong>, and <strong>Williams %R</strong>. Each indicator brings a unique perspective on price action and, when combined, provides a well-rounded approach to identify trading opportunities.
</p>

<h3 style="color:#2e86c1;">1. RSI (Relative Strength Index)</h3>
<p style="color:#2e86c1;">
  <strong>Purpose:</strong> RSI measures the strength of recent price movements, identifying overbought or oversold conditions.<br>
  <strong>Why it's useful:</strong> RSI highlights potential reversal points. Values below 30 suggest the asset is oversold, while values above 70 indicate it is overbought.<br>
  <strong>Why it's a good choice:</strong> As a momentum oscillator, RSI helps in catching trend reversals, adding valuable insights into market sentiment.
</p>

<h3 style="color:#2e86c1;">2. Bollinger Bands</h3>
<p style="color:#2e86c1;">
  <strong>Purpose:</strong> Bollinger Bands define price volatility by plotting two standard deviations away from a simple moving average.<br>
  <strong>Why it's useful:</strong> They highlight when prices are abnormally high or low relative to recent price action. When prices touch or breach the bands, it often signals an upcoming reversal.<br>
  <strong>Why it's a good choice:</strong> Combining Bollinger Bands with other indicators provides context about market volatility and helps confirm trend reversals or continuations.
</p>

<h3 style="color:#2e86c1;">3. MACD (Moving Average Convergence Divergence)</h3>
<p style="color:#2e86c1;">
  <strong>Purpose:</strong> MACD analyzes the relationship between two moving averages to gauge momentum and trend direction.<br>
  <strong>Why it's useful:</strong> MACD helps identify changes in trend momentum through crossovers between the MACD line and signal line.<br>
  <strong>Why it's a good choice:</strong> This indicator is excellent for confirming the strength of a trend, providing useful entry and exit points when combined with momentum-based indicators like RSI.
</p>

<h3 style="color:#2e86c1;">4. ATR (Average True Range)</h3>
<p style="color:#2e86c1;">
  <strong>Purpose:</strong> ATR measures the volatility of an asset by analyzing the range between its highest and lowest prices.<br>
  <strong>Why it's useful:</strong> ATR helps set stop-loss levels and identify periods of high or low volatility, ensuring better risk management.<br>
  <strong>Why it's a good choice:</strong> Including ATR in the strategy helps adjust risk parameters like stop-loss and take-profit levels, protecting against volatile price swings.
</p>

<h3 style="color:#2e86c1;">5. Williams %R</h3>
<p style="color:#2e86c1;">
  <strong>Purpose:</strong> Williams %R identifies overbought and oversold conditions by comparing the closing price to the high-low range over a set period.<br>
  <strong>Why it's useful:</strong> It provides early signals of potential trend reversals, often catching opportunities faster than other oscillators.<br>
  <strong>Why it's a good choice:</strong> Williams %R offers a complementary perspective to RSI, providing a more sensitive measure of overbought and oversold conditions.
</p>


<p style="color:#2e86c1;">
  <strong>Combining these Indicators:</strong> Each indicator has its strengths: momentum (RSI, Williams %R), trend (MACD), and volatility (Bollinger Bands, ATR). Combining them ensures that the strategy captures a variety of signals across different market conditions, improving accuracy and reducing the risk of false signals.
</p>
